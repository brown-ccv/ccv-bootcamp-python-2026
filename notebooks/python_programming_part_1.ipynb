  ```python
    # 1. Print the names of all the files in a directory individually.
    for file in files:
      print(file)

    # 2. How do we only print the csv files?
    for file in files:
      if file.endswith('.csv'):
        print(file)

    # 3. Can we add a counter to see how many csv’s are in the directory?
    counter = 0
    for file in files:
      if file.endswith('.csv'):
        print(file)
        counter = counter + 1
    print(f'Number of csv files: {counter}')
  ```

    * Output: csv file

    Extra credit: Also return a dictionary with the number of lines in each file

  ```python
  def combine_csv_files(path, suffix='.csv'):
      # List and sort files in the directory
      list_files_in_path = sorted(os.listdir(path))

      # Define output file and path
      output_filename = 'combined_file.csv'
      output_path = os.path.join(path, output_filename)

      # Initialize our counter
      file_count = 0

          # Open the output file for writing
      with open(output_path, 'w') as outfile:
              # Loop though each file
          for file in list_files_in_path:

              # Filter for the correct files
              if file.endswith(suffix) and file != output_filename:
                  file_path = os.path.join(path, file)

                  # Open the data file
                  with open(file_path, 'r') as infile:
                      lines = infile.readlines()

                      if not lines:  # Optional
                          continue

                      # Check if this is the very first file we are processing
                      if file_count == 0:
                          outfile.writelines(lines)       # Keep the header
                      else:
                          outfile.writelines(lines[1:])   # Skip the header

                      # Increment the counter so the next file hits the 'else' block
                      file_count += 1
  ```
